### **1.3. Operações**

As operações da estrutura do tipo *list* são:


*   `+` (concatenação).

**Exemplo**: Fabricantes de *hardware* mobile

```python
fabricante_mobile_chines = ['xiaomi', 'huawei']
fabricante_mobile_eua = ['apple', 'motorola']
fabricantes_mobile = fabricante_mobile_chines + fabricante_mobile_eua

print(fabricante_mobile_chines)
print(fabricante_mobile_eua)
print(fabricantes_mobile)
```

Outra operação muito utilizada é a de fatiamento (*slicing*), semelhante ao de *strings*:

Fatiamento fixo:

```python
print(f'0: {fabricantes_mobile[0]}')
print(f'- 1: {fabricantes_mobile[-1]}')
```

Fatiamento por intervalo:

```python
fabricante_mobile_chines = fabricantes_mobile[0:2]
fabricante_mobile_eua = fabricantes_mobile[2:len(fabricantes_mobile)]

print(f'China: ' + str(fabricante_mobile_chines))
print(f'EUA: ' + str(fabricante_mobile_eua))
```

Podemos adicionar elementos a uma posição específica da lista:

```python
print(fabricantes_mobile)
```

```python
fabricantes_mobile[2] = 'Nokia'
print(fabricantes_mobile)
```

Podemos acrescentar elementos a uma posição específica da lista:


```python
fabricantes_mobile.insert(2,'apple')
print(fabricantes_mobile)
```

### **1.4. Métodos**

São métodos nativos do Python que nos ajudam a trabalhar no dia a dia com listas.

```python
juros = [0.05, 0.07, 0.02, 0.04, 0.08]
print(juros)
```

```python
# inserir um elemento sem substituir: list.insert(index, val)
juros.insert(0, 0.33)
print(juros)
```

```python
# inserir um elemento no fim da lista: list.append(val)
juros.append(0.99)
print(juros)
```

```python
# remover um elemento pelo valor: list.remove(val)
juros.remove(0.77)
print(juros)
```

```python
# remover um elemento pelo índice: list.pop(val)
juros.pop(5)
print(juros)
```
