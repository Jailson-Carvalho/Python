## 1\. Estrutura condicional if / else / elif

Estrutura de alteração de fluxo lógico do código, avalia um valor booleano ou uma comparação lógica. **Note** a identação do código.

```python
if <booleano / comparação lógica> == True:
  <execute este código>
else:
  <senão execute este código>
```

````python
if False:
  print("Verdadeiro")
else:
  print("Falso")
````

**Exemplo**: Código de segurança de um cartão de crédito

````python
codigo_seguranca = '010'
codigo_seguranca_cadastrado = '010'

pode_efetuar_pagamento = codigo_seguranca == codigo_seguranca_cadastrado

print(pode_efetuar_pagamento)
````

````python
if codigo_seguranca == codigo_seguranca_cadastrado:
  print("Pode efetuar o pagamento.")
else:
  print("Erro! Código de segurança inválido!")
````

**Exemplo**: Código e senha de segurança de um cartão de crédito

````python
codigo_seguranca = str(input('Digite o código de segurança: '))
codigo_seguranca_cadastrado = '010'

senha = str(input('Digite a senha: '))
senha_cadastrada = '123'
````

````python
# operador AND &

if (codigo_seguranca == codigo_seguranca_cadastrado) & (senha == senha_cadastrada):
  print('Pode efetuar o pagamento!')

else:
  print('Código de segurança ou senha inválida!')
````

````python
# operador OR |
if (codigo_seguranca != codigo_seguranca_cadastrado) | (senha != senha_cadastrada):
  print('Código de segurança ou senha inválida!')

else:
  print('Pode efetuar o pagamento!')
````

### **1.2. if / elif / else**

Podemos também avaliar múltipla condições.

```python
if <1º booleano / 1ª comparação lógica> == True:
  <execute este código se a primeira condição for verdade>
elif <2º booleano / 2ª comparação lógica> == True:
  <execute este código se a segunda condição for verdade>
else:
  <senão execute este código>
```

````python
codigo_seguranca = str(input('Digite o código de segurança: '))
codigo_seguranca_cadastrado = '010'

senha = str(input('Digite a senha: '))
senha_cadastrada = '123'
````

| CÓDIGO        | SENHA        | | | CÓDIGO AND SENHA  | MENSAGEM |
|----------|----------|-|-|----------|-|
| **TRUE** | **TRUE** | | | **TRUE** | Pagamento efetuado |
| **TRUE** | FALSE    | | | FALSE    | Erro: Senha inválida |
| FALSE    | FALSE    | | | FALSE    | Erro: Código de segurança e senha inválidos |
| FALSE    | **TRUE** | | | FALSE    | Erro: Código de segurança inválido |

````python
if (codigo_seguranca == codigo_seguranca_cadastrado) & (senha == senha_cadastrada):
  print("Pode efetuar o pagamento.")

elif (codigo_seguranca == codigo_seguranca_cadastrado) & (senha != senha_cadastrada):
  print('Senha inválida.')

elif (codigo_seguranca != codigo_seguranca_cadastrado) & (senha == senha_cadastrada):
  print('Código de segurança inválido.')

else:
 print('Código de segurança e senha inválida.')
````

### **2.2. try / except**

Estrutura para tratar exceções:

````python
# Define o valor total da conta.
preco = 132.85
# Define a quantidade de pessoas que irão dividir o valor.
pessoas = 2

# O bloco try é usado para testar um código que pode gerar erros.
try:

  valor_por_pessoa = preco / pessoas# Calcula o valor que cada pessoa deverá pagar.
  print(f'Valor por pessoa é: R$ {valor_por_pessoa:.2f}') # Exibe o resultado formatado com duas casas decimais.

except ZeroDivisionError: # Captura especificamente o erro de divisão por zero.
  print('Número de pessoas inválido. Espera-se um valor maior que 0 e obteve-se um valor igual a ' +str(pessoas)) # Exibe uma mensagem informando que o número de pessoas é inválido.
````

````python
anos = [2023, 2024, 2025]
````

````python
try:
  ano_atual = anos[3]
  print('Ano atual')

# Caso ocorra qualquer erro, como acessar um índice inexistente, o programa executará o bloco 'except'
except Exception:
   # Exibe uma mensagem informando que a lista possui menos elementos do que o esperado e mostra o índice máximo disponível
  print('Lista de anos é menor que o valor escolhido. Espera-se um valor entre 0 e ' + str(len(anos) - 1))
````

### **2.3. try / except / finally**

````python
nome = 'Jailson Carvalho'
idade = 36
````

````python
# O bloco try tenta executar o código que cria a variável "apresentacao".
# Como a concatenação de strings com números gera um erro em Python,
# pode ocorrer uma exceção do tipo TypeError caso a variável "idade"
# seja um número inteiro.

# Se o erro acontecer, o bloco except é executado.
# Nele, a variável "idade" é convertida para string usando a função str(),
# permitindo que a concatenação seja realizada corretamente.

# O bloco finally é executado independentemente de ocorrer erro ou não.
# Ele exibe a mensagem "Segunda chance", recria a variável "apresentacao"
# com a idade já convertida em texto e imprime o resultado final.

try:
  apresentacao = 'Fala pessoal, meu nome é ' + nome + ' e eu tenho ' + idade + ' anos'
  print(apresentacao)
except TypeError:
  idade = str(idade)
finally:
  print('Segunda chance')
  apresentacao = 'Fala pessoal, meu nome é ' + nome + ' e eu tenho ' + idade + ' anos'
  print(apresentacao)
````

## 3\. Estrutura repetição for / in

### **3.1. for / in**

Estrutura que permite a execução repetida de um bloco de código repetidas vezes.

```python
for variavel_temporaria in coleção:
  <execute este código>
```

### **3.2. for / in / range**

Estrutura que permite a execução repetida de um bloco de código **n** vezes.

````python
for valor in range(6):
  print(valor)
````

````python
soma = 0
for valor in range(0, 100000):
  soma = soma + valor

print(soma)
````

````python
for multipo_dois in range(2, 10, 2): # iterar do 2 até o 9 pulando (step) de dois em dois.
  print(multipo_dois)
````

### **3.3. for / in / list**

Estrutura que permite a execução de um bloco de código para todos os elementos de uma lista.

````python
frutas = ['maca', 'banana', 'laranja', 'uva', 'pera']

for fruta in frutas:
  print(fruta)
````

````python
# Percorre cada caractere da frase e verifica se ele é 'A' ou 'z'.
# Se encontrar uma dessas letras, exibe uma mensagem na tela.

frase  = 'Olá pessoal! Meu nome é Jailson Carvalho'

for caracter in frase:
  if (caracter == "J") | (caracter == "C"):
    print(f"A letra '{caracter}' está presente na frase.")
````

### **3.4. for / in / dict**

Estrutura que permite a execução de um bloco de código para todos os elementos de um dicionário.

````python
credito = {'123': 750, '456': 812, '789': 980}
````

````python
for chave, valor in credito.items():
  print(f'Para o documento {chave}, o valor do escore de crédito é {valor}.')
  print('\n')
````

### **3.5. break / continue**

Estrutura que permite a quebra ou o avanço de um laço de repetição.

````python
for i in range(2, 10*10*10*10):
  print(i)
  if i == 10:
    break
````

````python
# Percorre a lista de números e verifica se cada valor é par ou ímpar.
# Se o resto da divisão por 2 for igual a zero, o número é par.
# Caso contrário, o número é ímpar.

numeros = [325, 333, 127, 999, 692, 222, 978, 487, 124, 156, 789 ,45, 3, 453, 782, 444, 365, 142, 365, 656, 4154]

for numero in numeros:

  if numero % 2 == 0:
    print(f'O némero {numero} é par!')
    break
  else:
    print(f'O número {numero} é impar!')
````

````python
# para imprimir todos

numeros = [325, 692, 487, 124, 156, 789 ,45, 3, 453, 782, 444, 365, 142, 365, 656, 4154]

print("Números pares")
for numero in numeros:
  if numero % 2 == 0:
    print(f'O némero {numero} é par!')

print('\n')

print("Números impares")
for numero in numeros:
  if numero % 2 != 0:
    print(f'O némero {numero} é impar!')
````
